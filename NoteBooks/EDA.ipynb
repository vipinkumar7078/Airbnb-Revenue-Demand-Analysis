```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
plt.style.use("ggplot")
```

```python
df = pd.read_excel("../data/ABNCleanexcel_data.xlsx")
```

```python
# basic info
df.head()
df.tail()
df.info()
df.describe()
df.shape
```

```python
# Price Distribution
plt.figure(figsize=(10,5))
plt.hist(df['price'], bins=50)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()

# Insight: Price kis range me zyada hai?
```

```python
# Room Type Distribution
room_counts = df['room_type'].value_counts()

plt.figure(figsize=(8,5))
room_counts.plot(kind='bar')
plt.title("Room Type Distribution")
plt.xlabel("Room Type")
plt.ylabel("Count")
plt.show()

```

```python
# City-wise Listings
city = df['city'].value_counts().head(10)

plt.figure(figsize=(10,5))
city.plot(kind='bar')
plt.title("Top 10 Cities by Listings")
plt.xlabel("City")
plt.ylabel("Listings")
plt.show()
```

```python
# avg price by city
avg_price = df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
avg_price.plot(kind='bar')
plt.title("Average Price by City")
plt.xlabel("City")
plt.ylabel("Average Price")
plt.show()
```

```python
# Availability Analysis
plt.figure(figsize=(10,5))
plt.hist(df['availability_365'], bins=30)
plt.title("Availability Distribution")
plt.xlabel("Availability")
plt.ylabel("Count")
plt.show()
```

```python
# top hots 
hosts = df['host_name'].value_counts().head(10)

plt.figure(figsize=(10,5))
hosts.plot(kind='bar')
plt.title("Top Hosts")
plt.xlabel("Host")
plt.ylabel("Listings")
plt.show()
```

```python
# Correlation Heatmap
numeric = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,8))
sns.heatmap(numeric.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#Scatter Plot
plt.figure(figsize=(8,5))
plt.scatter(df['price'], df['number_of_reviews'])
plt.xlabel("Price")
plt.ylabel("Number of Reviews")
plt.title("Price vs Reviews")
plt.show()
```

```python
# BOX plot
plt.figure(figsize=(10,5))
plt.boxplot(df['price'])
plt.title("Price Box Plot")
plt.show()
```
